```python
# load in the "starbucks.csv" dataset
# refer to how we read the cereal.csv dataset in the tutorial
import pandas as pd 
read_star = pd.read_csv("../data/starbucks.csv")
#type(read_star)
```

```python
# output the calories, sugars, and protein columns only of every 40th row. 
read_star.iloc[::40][["calories", "sugars", "protein"]]
```

```python
# select all rows with more than and including 400 calories
read_star[read_star["calories"] >= 400]
```

```python
# select all rows whose vitamin c content is higher than the iron content
read_star[read_star["vitamin c"] > read_star["iron"]]
```

```python
# create a new column containing the caffeine per calories of each drink
read_star["caff_per_cal"] = read_star["caffeine"] / read_star["calories"]
read_star.head()
```

```python
# what is the average calorie across all items?
read_star["calories"].mean()
```

```python
# how many different categories of beverages are there?
read_star["beverage"].nunique()
```

```python
# what is the average # calories for each beverage category?
bev_object = read_star.groupby("beverage")
bev_object["calories"].mean()
```

```python
# plot the distribution of the number of calories in drinks with a histogram
#cal_hist = read_star.groupby("beverage")["calories"].count()
read_star["calories"].plot.hist(edgecolor = "black", alpha = 0.8, title = "Calorie Distribution")
```

```python
# plot calories against total fat with a scatterplot
read_star.plot.scatter(x = "total fat", y = "calories", title = "Total Fat vs Calorie Distribution")
```
